```
X = img

if (X.shape[0] > X.shape[1]):
  result = np.zeros([X.shape[0],X.shape[0],4])
else:
  result = np.zeros([X.shape[1],X.shape[1],4])

def rotateBy90(x,y):
  iN = -1 * (y - (result.shape[0]-1)/2) + (result.shape[0]-1)/2 
  yN = (x)
  return int(iN),int(yN)

d = int(input("Enter degree in multiple of 90:"))
d = int(d / 90)
d = d % 4
if(d > 0):
  for i in range(X.shape[0]):
      for j in range(X.shape[1]):
        x, y = rotateBy90(i,j)
        for k in range(d-1):
          x, y = rotateBy90(x, y)
        result[x,y] = X[i][j]
  cv2_imshow(result)
else:
  cv2_imshow(X)
```
