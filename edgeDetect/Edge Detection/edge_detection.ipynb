```
import numpy as np
import cv2
from PIL import Image
from google.colab.patches import cv2_imshow
```

```
img = np.array(Image.open('/content/drive/My Drive/Colab Notebooks/edge_detection/edge-detection2.jpg'))
cv2_imshow(img)
```

```


result = np.zeros((img.shape[0]+2,img.shape[1]+2,img.shape[2]))

result[1 :-1,1:-1,:] = img
```

```
vertical_sober = np.array([
                           [-1,-2,-1],
                           [0,0,0],
                           [1,2,1]
                          ])

horizontal_sober = [[-1,0,1], 
                     [-2,0,2], 
                     [-1,0,1]]
```

```
sharped = np.zeros(img.shape)

def multiplier():
  for i in range(img.shape[0]):
    for j in range(img.shape[1]):
      imp = result[i: i+3, j: j+3, 0]
      
      ver_pixels = vertical_sober*imp

      ver_score = ver_pixels.sum()/4

      hor_pixels = horizontal_sober*imp

      hor_score = hor_pixels.sum()/4

      transform_score = (ver_score**2 + hor_score**2)**.5
      
      sharped[i,j] = [transform_score]*3
      

multiplier()
cv2_imshow(sharped)
```
